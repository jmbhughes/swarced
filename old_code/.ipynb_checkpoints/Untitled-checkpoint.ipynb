```python
r1u = analyze(getQuery("202060551","0"))
```

```python
r1u.response['peaks']
```

```python
period=r1u.peaks[0]['period']
t0 = r1u.peaks[0]['t0']
for lc in r1u.model_light_curves:
    mu = lc.predict()
    t = (lc.time - t0 + 0.5 * period) % period - 0.5 * period
    pl.scatter(t, lc.flux-mu,c='black');
pl.show()
```

```python
for lc in r1u.model_light_curves:
    pl.xlabel("time [days]")
    pl.ylabel("de-trended aperture flux [ppt]")
    pl.plot(lc.time, lc.flux - lc.predict(), 'k.')
    #pl.plot(lc.time,lc.predict(), 'g.')
pl.show()
```
