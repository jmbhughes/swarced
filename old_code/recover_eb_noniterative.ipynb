We want to use the Foreman-Mackey code possibly to extract eclipsing binaries. I took a list of known eclisping binaries from the Kepler EBS working group and ran it through ketu. Below you see the results. Please note this uses the full 2-d search instead of the iterative one. 

```python
from __future__ import division, print_function

%matplotlib inline
%config InlineBackend.figure_format = "retina"

from matplotlib import rcParams
rcParams["savefig.dpi"] = 100
rcParams["font.size"] = 20
import h5py
import transit
import numpy as np
import matplotlib.pyplot as pl
from matplotlib.ticker import MaxNLocator
import ketu
import urllib2
import sys
from astropy.io import fits
from astropy.io import ascii
#from eliminate_eclipse import *
import pipeline
from eliminate_eclipses import *
```

These functions likley have an updated version in the swarced file

```python
def load(epicid):
    un = pipeline.analyze_full(pipeline.getQuery(epicid,"0"))
    cl = pipeline.analyze_full(pipeline.getEBQuery(epicid,"0"))
    return un, cl
```

```python
def analyzer(epicid):
    pipe = ketu.k2.Data(basepath="/k2_data/eb_removed/cache/")
    pipe = ketu.k2.Likelihood(pipe)
    pipe = ketu.OneDSearch(pipe)
    pipe = ketu.TwoDSearch(pipe)
    pipe = ketu.PeakDetect(pipe)
    result = pipe.query(**pipeline.getQuery(str(epicid),"0"))
    return result
```

```python
def plotperiodogram(epicid):
    #This plots the more reliable phic periodogram as calculated in the full 2-d search
    result = analyze(epicid)
    fig = pl.figure(figsize=(10, 4))
    ax = fig.add_subplot(111)
    parent_response = result.response
    peaks = parent_response['peaks']
    x = parent_response['periods']
    y = parent_response['phic_scale']/100
    m = np.isfinite(y)
    ax.plot(x[m], y[m], "k")
    pl.xticks(np.arange(2,30,2))
    pl.xlim([0,30])
    for i, peak in enumerate(peaks):
        x0, y0 = peak["period"], peak["phic_norm"]/100
        ax.plot(x0, y0, ".r")
        ax.annotate("{0:.2f}".format(x0), xy=(x0, y0), ha="center",fontsize=10,
                    xytext=(10, 5), textcoords="offset points")
```

```python
def plots2n(epicid):
    #This plots signal to noise periodogram as calculated in the iterative step see equations 6 and 7 in the paper
    result = analyzer(epicid)
    phic_same = result.parent_response.response['phic_same']
    t0_2d = result.parent_response.response['t0_2d']
    depth_2d = result.parent_response.response['depth_2d']
    depth_ivar_2d = result.parent_response.response['depth_ivar_2d']
    periods = result.parent_response.response['period_2d']
    query = pipeline.getQuery(str(epicid),0)
    # Profile over duration.
    inds = np.arange(len(phic_same)), np.argmax(phic_same, axis=1)
    t0_2d = t0_2d[inds]
    depth_2d = depth_2d[inds]
    depth_ivar_2d = depth_ivar_2d[inds]

    # Find the top peak.
    s2n = depth_2d * np.sqrt(depth_ivar_2d)
    top_peak = np.argmax(s2n)
    p, t0 = periods[top_peak], t0_2d[top_peak]
    duration = query["durations"][inds[1][top_peak]]
    
    fig = pl.figure(figsize=(10, 4))
    ax = fig.add_subplot(111)
    parent_response = result.response
    peaks = parent_response['peaks']
    x = parent_response['periods']
    y = s2n
    m = np.isfinite(y)
    ax.plot(x[m], y[m], "k")
    pl.xticks(np.arange(2,30,2))
    pl.xlim([0,30])
    ax.plot(p, s2n[top_peak], ".r")
    ax.annotate("{0:.2f}".format(p), xy=(p, s2n[top_peak]), ha="center",fontsize=10,
                    xytext=(10, 5), textcoords="offset points")
```

Here I just run ove rthe list of EBs and store the results in array l

```python
l = np.array([0,0,0,0])
for i in range(len(r)):
    epicid = r['KIC/EPIC'][i]
    result = analyzer(epicid)
    l = np.vstack((l,np.array([r['period'][i], result.response['peaks'][0]['period'], 
                               result.response['peaks'][0]['phic_norm'],
                              result.response['peaks'][0]['depth_s2n']])))
l = l[1:]
```

```python
kepler_period, recovered_period = r['period'], l[:,1]
s2n = l[:,3]
```
